IRIS DataSet


Task 1: Import and Explore Data

```python
import numpy  as np
import pandas as pd
import seaborn as sns
iris = sns.load_dataset("iris")

# Display top 5 rows
iris.head()

# Display bottom 5 rows
iris.tail()

# Show column info
iris.info()

# Show statistical summary
iris.describe()
```

Task 2: Sorting and Filtering

```python
iris.sort_values(by="petal_length", ascending=True)
iris[iris["sepal_width"] > 3.5]
iris.query("species == 'setosa'")


```

âœ… Task 3: Slicing and Transposing

```python
iris_subset = iris.loc[:9, ["sepal_length", "species"]]
iris.T
dummy = pd.DataFrame({
    "sepal_length": [5.1, 4.9, 5.0, 5.5, 6.0],
    "sepal_width": [3.5, 3.0, 3.3, 3.8, 2.9],
    "petal_length": [1.4, 1.6, 1.5, 1.2, 4.5],
    "petal_width": [0.2, 0.4, 0.3, 0.2, 1.5],
    "species": ["setosa"]*5
})

iris_appended = pd.concat([iris, dummy], ignore_index=True)
iris_truncated = iris_appended.head(100)
print(iris_truncated)
```

```python
tips = sns.load_dataset("tips")

tips.shape      # Rows & columns
tips.info()
tips.describe()
```

```python
tips.sort_values("total_bill", ascending=False)
tips[tips["tip"] > 5]
tips.sort_values(["day", "time"])
```

```python
slice_tips = tips.loc[:14, ["total_bill", "tip", "sex"]]
slice_tips.T #transposing
```
