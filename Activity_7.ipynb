```python
def check_popularity(pop):
    if pop > 70:
        return "High"
    elif pop >= 40:
        return "Medium"
    else:
        return "Low"

spotify_df["popularity_level"] = spotify_df["popularity"].apply(check_popularity)
print(spotify_df.head())
```

task# 13

```python
import xml.etree.ElementTree as ET

root = ET.Element("spotify")

for _, row in spotify_df.iterrows():
    song = ET.SubElement(root, "song")

    for col in spotify_df.columns:
        child = ET.SubElement(song, col)
        child.text = str(row[col])

tree = ET.ElementTree(root)
tree.write("spotify_data.xml")

print("spotify_data.xml created!")
```
