```python
#Task 7
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ali", "Sara", "John", "Ayesha"],
    "Average": [95, 88, 92, 85]
})

df2 = pd.DataFrame({
    "Student": ["Ali", "Sara", "John", "Ayesha"],
    "Attendance": [95, 98, 90, 100]
})

merged = pd.merge(df, df2, left_on="Name", right_on="Student")

merged = merged.drop(columns="Student")

print("Combined Table:")
print(merged)
```

```python
#Task 8
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ali", "Sara", "John", "Ayesha"],
    "Average": [95, 88, 92, 85]
})

df2 = pd.DataFrame({
    "Student": ["Ali", "Sara", "John", "Ayesha"],
    "Attendance": [95, 98, 90, 100]
})

merged = pd.merge(df, df2, left_on="Name", right_on="Student").drop(columns="Student")

merged["Final_Score"] = (merged["Average"] * 0.8) + (merged["Attendance"] * 0.2)

print("Combined Table with Final_Score:")
print(merged)
```
